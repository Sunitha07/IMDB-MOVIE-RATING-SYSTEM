```python
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Assuming you have your feature matrix stored in 'scaled_data' and target variable in 'target'

# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(scaled_data, target, test_size=0.25, random_state=42)

# Train the SVM model
svm_regressor = SVR(kernel='rbf')  # You can specify other kernels as well
svm_regressor.fit(x_train, y_train)

# Make predictions on the test set
svm_predictions = svm_regressor.predict(x_test)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, svm_predictions)
print('SVM Mean Absolute Error:', round(mae, 2), 'degrees.')

# Calculate Accuracy
errors = abs(svm_predictions - y_test)
mape = 100 * (errors / y_test)
accuracy = 100 - np.mean(mape)
print('SVM Accuracy:', round(accuracy, 2), '%.')

# Create a DataFrame with the actual and predicted values
predictions = pd.DataFrame({"Actual": y_test, "Predicted": svm_predictions})
print(predictions)
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# Assuming you have your feature matrix stored in 'scaled_data' and target variable in 'target'

# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(scaled_data, target, test_size=0.25, random_state=42)

# Train the Lasso Regression model
lasso_regressor = Lasso(alpha=1.0)  # You can adjust the alpha parameter as needed
lasso_regressor.fit(x_train, y_train)

# Make predictions on the test set
lasso_predictions = lasso_regressor.predict(x_test)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, lasso_predictions)
print('Lasso Regression Mean Absolute Error:', round(mae, 2), 'degrees.')

# Create a DataFrame with the actual and predicted values
predictions = pd.DataFrame({"Actual": y_test, "Predicted": lasso_predictions})
print(predictions)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_absolute_error
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# Assuming you have your feature matrix stored in 'scaled_data' and target variable in 'target'

# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(scaled_data, target, test_size=0.25, random_state=42)

# Convert data into DMatrix format for XGBoost
dtrain = xgb.DMatrix(x_train, label=y_train)
dtest = xgb.DMatrix(x_test, label=y_test)

# Parameters for XGBoost
params = {
    'objective': 'reg:squarederror',  # Use regression objective
    'eval_metric': 'mae'  # Mean Absolute Error as evaluation metric
}

# Training the XGBoost model
num_rounds = 100  # Number of boosting rounds
xgb_model = xgb.train(params, dtrain, num_rounds)

# Make predictions on the test set
xgb_predictions = xgb_model.predict(dtest)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, xgb_predictions)
print('XGBoost Mean Absolute Error:', round(mae, 2), 'degrees.')

# Create a DataFrame with the actual and predicted values
predictions = pd.DataFrame({"Actual": y_test, "Predicted": xgb_predictions})
print(predictions)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_absolute_error
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# Assuming you have your feature matrix stored in 'scaled_data' and target variable in 'target'

# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(scaled_data, target, test_size=0.25, random_state=42)

# Train the AdaBoost model
adaboost_regressor = AdaBoostRegressor(n_estimators=50, random_state=42)  # You can adjust the number of estimators as needed
adaboost_regressor.fit(x_train, y_train)

# Make predictions on the test set
adaboost_predictions = adaboost_regressor.predict(x_test)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, adaboost_predictions)
print('AdaBoost Mean Absolute Error:', round(mae, 2), 'degrees.')

# Create a DataFrame with the actual and predicted values
predictions = pd.DataFrame({"Actual": y_test, "Predicted": adaboost_predictions})
print(predictions)
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import mean_absolute_error
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# Assuming you have your feature matrix stored in 'scaled_data' and target variable in 'target'

# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(scaled_data, target, test_size=0.25, random_state=42)

# Train the CatBoost model
catboost_regressor = CatBoostRegressor(iterations=1000,  # Number of trees (iterations)
                                       learning_rate=0.1,  # Learning rate
                                       depth=6,  # Depth of trees
                                       loss_function='MAE',  # Loss function: Mean Absolute Error
                                       random_seed=42)  # Random seed for reproducibility

catboost_regressor.fit(x_train, y_train, verbose=False)  # verbose=False to suppress output

# Make predictions on the test set
catboost_predictions = catboost_regressor.predict(x_test)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, catboost_predictions)
print('CatBoost Mean Absolute Error:', round(mae, 2), 'degrees.')

# Create a DataFrame with the actual and predicted values
predictions = pd.DataFrame({"Actual": y_test, "Predicted": catboost_predictions})
print(predictions)
```
